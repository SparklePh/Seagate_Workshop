```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Read the image and convert to RGB
img = Image.open("data/cat.jpg").convert("RGB")
img_array = np.array(img)

# Plot the image
plt.figure(figsize=(12, 8))
plt.imshow(img_array)
plt.axis("on")  # Turn on the axis to show X and Y coordinates

# Set ticks every 50 pixels
plt.xticks(np.arange(0, img_array.shape[1], 50))
plt.yticks(np.arange(0, img_array.shape[0], 50))

# Add grid lines
plt.grid(color='white', linestyle='--', linewidth=0.5)  # White dashed grid

# Add axis labels and title
plt.xlabel("X (pixels)")
plt.ylabel("Y (pixels)")
plt.title("Image with pixel grid")

# Display the plot
plt.show()
```

# Grayscale channel Vs RGB channel



```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Read the image and convert to grayscale
img_pil = Image.open("data/cat.jpg").convert("L")  # 'L' = grayscale
img_array = np.array(img_pil)

# Print shape and mode
print("shape:", img_array.shape)   # (height, width) for grayscale
print("mode:", img_pil.mode)       # 'L' for grayscale

# Plot the grayscale image
plt.figure(figsize=(12, 8))
plt.imshow(img_array, cmap='gray')
plt.axis("on")
plt.xlabel("X (pixels)")
plt.ylabel("Y (pixels)")
plt.title("Grayscale Image")
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# อ่านภาพ
img_pil = Image.open("data/cat.jpg")
img_array = np.array(img_pil)

print("shape:", img_array.shape)   # (height, width, channels)
print("mode:", img_pil.mode)       # RGB
```

```python
R = img_array[:,:,0]

print(f"R.shape : {R.shape}")
print(f"array:{R}")

```

```python
G = img_array[:,:,1]

print(f"R.shape : {G.shape}")
print(f"array:{G}")
```

```python
B = img_array[:,:,2]

print(f"R.shape : {B.shape}")
print(f"array:{B}")
```

```python
# แสดงผล
fig, axs = plt.subplots(1, 4, figsize=(15,5))

axs[0].imshow(img_pil)
axs[0].set_title("Original")

axs[1].imshow(R, cmap="Reds")
axs[1].set_title("Red Channel")

axs[2].imshow(G, cmap="Greens")
axs[2].set_title("Green Channel")

axs[3].imshow(B, cmap="Blues")
axs[3].set_title("Blue Channel")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img_cv = cv2.imread("data/cat.jpg")
print("shape:", img_cv.shape)

# แสดงผล
plt.imshow(img_cv)
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread("data/cat.jpg")
print("shape:", img_bgr.shape)   #(427, 640, 3)

# แยก BGR ออกมา
B, G, R = cv2.split(img_bgr)

# Merge channels back to original
img_merged = cv2.merge([B, G, R])

# แสดงผล
fig, axs = plt.subplots(1, 5, figsize=(15,5))

axs[0].imshow(img_bgr)
axs[0].set_title("OpenCV")
axs[0].axis("off")

axs[1].imshow(B, cmap="Blues")
axs[1].set_title("Blue channel")
axs[1].axis("off")

axs[2].imshow(G, cmap="Greens")
axs[2].set_title("Green channel")
axs[2].axis("off")

axs[3].imshow(R, cmap="Reds")
axs[3].set_title("Red channel")
axs[3].axis("off")

axs[4].imshow(img_merged)
axs[4].set_title("img_bgr_merged")
axs[4].axis("off")

plt.show()
```

# แบบฝึกหัด 2
> อยากได้น้องแมวปกติ RGB ต้องทำไง


```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread("data/cat.jpg")
print("shape:", img_bgr.shape)   #(427, 640, 3)

# แยก BGR ออกมา
B, G, R = cv2.split(img_bgr)

# Merge channels back to original
img_merged = cv2.merge([_, _, _])

# แสดงผล
fig, axs = plt.subplots(1, 5, figsize=(15,5))

axs[0].imshow(img_bgr)
axs[0].set_title("OpenCV")
axs[0].axis("off")

axs[1].imshow(B, cmap="Blues")
axs[1].set_title("Blue channel")
axs[1].axis("off")

axs[2].imshow(G, cmap="Greens")
axs[2].set_title("Green channel")
axs[2].axis("off")

axs[3].imshow(R, cmap="Reds")
axs[3].set_title("Red channel")
axs[3].axis("off")

axs[4].imshow(img_merged)
axs[4].set_title("img_bgr_merged")
axs[4].axis("off")

plt.show()
```

# มี tool ช่วยแปลง ภาพจาก BGR -> RGB ให้

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img_bgr = cv2.imread("data/cat.jpg")

# Convert BGR to RGB for correct display in matplotlib
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

print("shape:", img_rgb.shape)

# แสดงผล
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```
